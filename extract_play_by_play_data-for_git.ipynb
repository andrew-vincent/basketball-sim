```python
# Import libraries
import pandas as pd
import numpy as np
import random
import scipy
import math
import requests as requests
from bs4 import BeautifulSoup 
from datetime import datetime
```

```python
# Get Game Play-by-play HTML file
web_url = 'https://www.basketball-reference.com/boxscores/pbp/202305070PHI.html'
fetched_page = requests.get(web_url)
soup = BeautifulSoup(fetched_page.text)
game_info = soup.h1(text=True)
plays = soup.find_all('td')
```

```python
# Convert plays elements to strings
i = 0
while i < len(plays):
    plays[i] = plays[i].get_text()
    i+=1
    
# Remove empty lines
plays = [s for s in plays if s != '\xa0']
```

```python
# Separate the Quarters into arrays
# Need to write code for OTs
arr1 = np.array(plays)
idx1 = np.where(arr1 == 'End of 1st quarter')[0]
subarrays1 = np.split(arr1, idx1+1)
result1 = [subarray.tolist() for subarray in subarrays1 if len(subarray) > 0]
Q1_plays = result1[0]

arr2 = np.array(result1[1])
idx2 = np.where(arr2 == 'End of 2nd quarter')[0]
subarrays2 = np.split(arr2, idx2+1)
result2 = [subarray.tolist() for subarray in subarrays2 if len(subarray) > 0]
Q2_plays = result2[0]

arr3 = np.array(result2[1])
idx3 = np.where(arr3 == 'End of 3rd quarter')[0]
subarrays3 = np.split(arr3, idx3+1)
result3 = [subarray.tolist() for subarray in subarrays3 if len(subarray) > 0]
Q3_plays = result3[0]

arr4 = np.array(result3[1])
idx4 = np.where(arr4 == 'End of 4th quarter')[0]
subarrays4 = np.split(arr4, idx4+1)
result4 = [subarray.tolist() for subarray in subarrays4 if len(subarray) > 0]
Q4_plays = result4[0]

if len(Q4_plays) > 0:
    try:
        arrOT1 = np.array(result4[1])
        idxOT1 = np.where(arrOT1 == 'End of 1st overtime')[0]
        subarraysOT1 = np.split(arrOT1, idxOT1+1)
        resultOT1 = [subarray.tolist() for subarray in subarraysOT1 if len(subarray) > 0]
        OT1_plays = resultOT1[0]
        if len(OT1_plays) > 0:
            try:
                arrOT2 = np.array(resultOT1[1])
                idxOT2 = np.where(arrOT2 == 'End of 2nd overtime')[0]
                subarraysOT2 = np.split(arrOT2, idxOT2+1)
                resultOT2 = [subarray.tolist() for subarray in subarraysOT2 if len(subarray) > 0]
                OT2_plays = resultOT2[0]
                if len(OT2_plays) > 0:
                    try:
                        arrOT3 = np.array(resultOT2[1])
                        idxOT3 = np.where(arrOT3 == 'End of 3rd overtime')[0]
                        subarraysOT3 = np.split(arrOT3, idxOT3+1)
                        resultOT3 = [subarray.tolist() for subarray in subarraysOT3 if len(subarray) > 0]
                        OT3_plays = resultOT3[0]
                    except:
                        print('No OT3')
            except:
                print('No OT2')
    except IndexError:
        print('No OT1')
```

```python
# Create arrays for each play separated by time change on the clock
times_indices1 = []
for x in Q1_plays:
    if x.find(':') != -1:
        colon_idx = x.find(':')
        next_char = x[colon_idx+1]
        if next_char.isnumeric() == True:
            times_indices1.append((Q1_plays.index(x)))

times_indices1 = sorted(list(set(times_indices1)))
Q1_split_plays = np.split(Q1_plays, times_indices1)

times_indices2 = []
for x in Q2_plays:
    if x.find(':') != -1:
        colon_idx = x.find(':')
        next_char = x[colon_idx+1]
        if next_char.isnumeric() == True:
            times_indices2.append((Q2_plays.index(x)))

times_indices2 = sorted(list(set(times_indices2)))
Q2_split_plays = np.split(Q2_plays, times_indices2)

times_indices3 = []
for x in Q3_plays:
    if x.find(':') != -1:
        colon_idx = x.find(':')
        next_char = x[colon_idx+1]
        if next_char.isnumeric() == True:
            times_indices3.append((Q3_plays.index(x)))

times_indices3 = sorted(list(set(times_indices3)))
Q3_split_plays = np.split(Q3_plays, times_indices3)

times_indices4 = []
for x in Q4_plays:
    if x.find(':') != -1:
        colon_idx = x.find(':')
        next_char = x[colon_idx+1]
        if next_char.isnumeric() == True:
            times_indices4.append((Q4_plays.index(x)))

times_indices4 = sorted(list(set(times_indices4)))
Q4_split_plays = np.split(Q4_plays, times_indices4)

# Overtime 1
try:
    times_indices_OT1 = []
    for x in OT1_plays:
        if x.find(':') != -1:
            colon_idx = x.find(':')
            next_char = x[colon_idx+1]
            if next_char.isnumeric() == True:
                times_indices_OT1.append((OT1_plays.index(x)))

    times_indices_OT1 = sorted(list(set(times_indices_OT1)))
    OT1_split_plays = np.split(OT1_plays, times_indices_OT1)
except NameError:
    pass

# Overtime 2
try:
    times_indices_OT2 = []
    for x in OT2_plays:
        if x.find(':') != -1:
            colon_idx = x.find(':')
            next_char = x[colon_idx+1]
            if next_char.isnumeric() == True:
                times_indices_OT2.append((OT2_plays.index(x)))

    times_indices_OT2 = sorted(list(set(times_indices_OT2)))
    OT2_split_plays = np.split(OT1_plays, times_indices_OT2)
except NameError:
    pass

#Overtime 3
try:
    times_indices_OT3 = []
    for x in OT3_plays:
        if x.find(':') != -1:
            colon_idx = x.find(':')
            next_char = x[colon_idx+1]
            if next_char.isnumeric() == True:
                times_indices_OT3.append((OT3_plays.index(x)))

    times_indices_OT3 = sorted(list(set(times_indices_OT3)))
    OT3_split_plays = np.split(OT3_plays, times_indices_OT3)
except NameError:
    pass
```

```python
# Q1 Shots from Plays DataFrame
game_shots_df_v2 = game_plays_v2[game_plays_v2["Type"].isin(["Shot"])].reset_index()

# Make or Miss flag
make_or_miss = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('makes') != -1:
        make_or_miss.append(True)
        i +=1
    else:
        make_or_miss.append(False)
        i += 1
game_shots_df_v2['Make or Miss'] = make_or_miss

# Pull Shooter name
shooter = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('makes') != -1:
        shooter.append(game_shots_df_v2['Play'][i][0:game_shots_df_v2['Play'][i].find('makes')-1])
        i +=1
    else:
        shooter.append(game_shots_df_v2['Play'][i][0:game_shots_df_v2['Play'][i].find('misses')-1])
        i += 1
game_shots_df_v2['Shooter'] = shooter

# Available points on shot
pts_avail = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('2-pt') != -1:
        pts_avail.append(2)
        i +=1
    elif game_shots_df_v2['Play'][i].find('3-pt') != -1:
        pts_avail.append(3)
        i +=1
    else:
        pts_avail.append(1)
        i +=1
game_shots_df_v2['Points Available'] = pts_avail
        
# Shot type
shot_type = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('flagrant free throw') != -1:
        shot_type.append('Flagrant Free Throw')
        i +=1
    elif game_shots_df_v2['Play'][i].find('free throw') != -1:
        shot_type.append('Free Throw')
        i +=1
    elif game_shots_df_v2['Play'][i].find('jump shot') != -1:
        shot_type.append('Jump Shot')
        i +=1
    elif game_shots_df_v2['Play'][i].find('layup') != -1:
        shot_type.append('Layup')
        i +=1
    elif game_shots_df_v2['Play'][i].find('hook shot') != -1:
        shot_type.append('Hook Shot')
        i +=1
    elif game_shots_df_v2['Play'][i].find('dunk') != -1:
        shot_type.append('Dunk')
        i +=1
    else:
        shot_type.append('Error')
        i +=1
game_shots_df_v2['Shot Type'] = shot_type


# Distance
dist = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('ft') != -1:
        dist.append(int(game_shots_df_v2['Play'][i][game_shots_df_v2['Play'][i].find('ft')-3:game_shots_df_v2['Play'][i].find('ft')-1].strip(' ')))
        i +=1
    elif game_shots_df_v2['Play'][i].find('at rim') != -1:
        dist.append(0)
        i +=1
    else:
        dist.append(15)
        i +=1

game_shots_df_v2['Distance'] = dist


# Shot Assisted
assisted = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('assist') != -1:
        assisted.append(True)
        i +=1
    else:
        assisted.append(False)
        i +=1
game_shots_df_v2['Assisted'] = assisted

#Assisted By
assisted_by = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('assist') != -1:
        assisted_by.append(game_shots_df_v2['Play'][i][game_shots_df_v2['Play'][i].find('assist')+10:-1])
        i +=1
    else:
        assisted_by.append('N/A')
        i +=1
game_shots_df_v2['Assisted By'] = assisted_by

# Shot Blocked
blocked = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('block') != -1:
        blocked.append(True)
        i +=1
    else:
        blocked.append(False)
        i +=1
game_shots_df_v2['Blocked'] = blocked

#Blocked By
blocked_by = []
i = 0
while i < len(game_shots_df_v2['Play']):
    if game_shots_df_v2['Play'][i].find('block') != -1:
        blocked_by.append(game_shots_df_v2['Play'][i][game_shots_df_v2['Play'][i].find('block')+9:-1])
        i +=1
    else:
        blocked_by.append('N/A')
        i +=1
game_shots_df_v2['Blocked By'] = blocked_by


#Quarter
quarter = []
i = 0
while i < len(game_shots_df_v2['Play']):
    quarter.append(game_shots_df_v2['Quarter'][i])
    i += 1
game_shots_df_v2['Quarter'] = quarter


game_shots_df_v2   
```
